```python
# Linear graph

import matplotlib.pyplot as plt

plt.title("Basic Linear Graph")
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
#Graph to plot distance and time

distance=list(range(2,11,2))
time=list(range(1,6))
#d=2*t

plt.title("Distance vs Time")   #Independent entity time taken on X axis
plt.xlabel("Time")
plt.ylabel("distance")
plt.plot(time,distance,'ro')  # in 'ro' , r is red color, 'bo'--> blue ,'go'-->green
plt.grid()
plt.show()
```

```python
#v=at
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i  for i in range(len(t))]

plt.title("Non Linear Graph")
plt.plot(t,v,"go")
plt.grid()
plt.show()
```

```python
n=list(range(10))
sqn=[i**2 for i in n]
plt.plot(n,sqn)
plt.grid()
```

```python
# voltage-Time

import numpy as np

#t=[i/10 for i in range(1,21)]

t=np.arange(0.1,2.0,0.1)

v=[1+np.sin(2*np.pi*i) for i in t]

plt.plot(t,v,'bo')
plt.grid()
plt.title('Time vs Voltage Graph')
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('Images/voltage-time.png')
plt.show()
```

```python
#Plot income of Arkansas from 2005 to 2013

import matplotlib.pyplot as plt


import pandas as pd
df=pd.read_csv('DataFiles/income.csv')
years=df.columns[2:]

incomeArkansas=df.values[3,2:]

plt.title("Income of Arkansas")
plt.xlabel('years')
plt.ylabel('income')
plt.plot(years,incomeArkansas,'bo')
plt.show()
```

```python
#income of all states and years

years=df.columns[2:]

incomeallstates=[]

for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
    
plt.plot(years,incomeallstates)
plt.title('All states Income')
plt.xlabel('years')
plt.ylabel('incomeallstates')
```
